### Meta

```python
# Importing mask
meta_mask = np.array(Image.open('./meta_logo.png'))
```

```python
meta_mask
```

```python
def transform_zeros(val):
    if val == 0:
        return 255
    else:
        return val
```

```python
maskable_image = np.ndarray((meta_mask.shape[0],meta_mask.shape[1]), np.int32)
```

```python
for i in range(len(meta_mask)):
    maskable_image[i] = list(map(transform_zeros, meta_mask[i]))
```

```python
# Plot the wordcloud with the mask applied
wc = WordCloud(background_color='white', colormap = 'binary', stopwords = ['meta'], width = 800, height = 500).generate(meta_text)
plt.figure(figsize=[16,9])
plt.axis("off")
plt.imshow(wc)
plt.show()
```

```python
# Plot the wordcloud with the mask applied
wc = WordCloud(background_color='black', mask= maskable_image, contour_width=2, contour_color='black', colormap = 'BuPu_r').generate(meta_text)
plt.figure(figsize=[16,16])
plt.tight_layout()
plt.imshow(wc)
plt.axis("off")
plt.show()
```
